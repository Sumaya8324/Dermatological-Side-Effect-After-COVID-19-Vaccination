```python
import pandas as pd

# Clean column names
df.columns = df.columns.str.strip()

# Categorical columns
categorical_columns = [
    'Age of the participants',
    'Gender of the participants',
    'Religion of the participants',
    'Location of the participants',
    'Nationality of the participants',
    'Division of the participants',
    'Occupation of the participants',
    'Education of the participants',
    'Health Status of the participants'
]

# Frequency and percentage table
for col in categorical_columns:
    print(f"\nVariable: {col}")
    freq_table = df[col].value_counts(dropna=False).reset_index()
    freq_table.columns = [col, 'Frequency']
    freq_table['Percent'] = (freq_table['Frequency'] / freq_table['Frequency'].sum()) * 100
    print(freq_table.to_string(index=False, formatters={'Percent': '{:,.1f}%'.format}))
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('covid.csv')


vaccine_data = {
    'Vaccine Name': ["Don't Know The Name", "Sinovac", "Sinopharm", "Moderna", "Pfizer", "AstraZeneca"],
    'Rural': [80, 20, 60, 70, 90, 50],
    'Urban': [120, 10, 40, 50, 110, 30]
}

# Create a DataFrame
df = pd.DataFrame(vaccine_data)

# Set the positions and width for the bars
bar_width = 0.35
index = range(len(df['Vaccine Name']))

# Plotting the grouped bar chart
fig, ax = plt.subplots(figsize=(10, 6))

bar1 = ax.bar(index, df['Rural'], bar_width, label='Rural', color='#2c3e50')
bar2 = ax.bar([i + bar_width for i in index], df['Urban'], bar_width, label='Urban', color='#27ae60')

# Customize the plot
ax.set_xlabel('Vaccine Name')
ax.set_ylabel('Number of Vaccinations')
ax.set_title('Distribution of Vaccines over Residence Type')
ax.set_xticks([i + bar_width/2 for i in index])
ax.set_xticklabels(df['Vaccine Name'], rotation=45, ha='right')
ax.legend()

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Display the plot
plt.savefig('residence_bar_chart.png')
plt.show()
```

```python
import pandas as pd
import scipy.stats as stats

# Clean column names (if not already done)
df.columns = df.columns.str.strip()

# Target variable (must match exactly your column name)
target = 'Did you experience any skin-related symptoms after receiving the COVID-19 vaccine?'  # Yes/No

# List of independent categorical variables
independent_vars = [
    'Age of the participants',
    'Gender of the participants',
    'Religion of the participants',
    'Location of the participants',
    'Education of the participants',
    'Occupation of the participants',
    'Health Status of the participants',
    'Which vaccine did you receive?',
    'How many doses have you received?'
]

# Loop through each variable and do crosstab + chi-square test
for var in independent_vars:
    print(f"\nVariable: {var}")
    
    # Crosstab
    ct = pd.crosstab(df[var], df[target], margins=False)

    # Percent row-wise (optional)
    percent_ct = ct.div(ct.sum(axis=1), axis=0) * 100
    
    # Combine frequency and percentage nicely
    formatted_table = ct.astype(str) + ' (' + percent_ct.round(1).astype(str) + '%)'
    print(formatted_table)

    # Chi-square test
    chi2, p, dof, expected = stats.chi2_contingency(ct)
    print(f"P-value: {p:.4f}")
```
